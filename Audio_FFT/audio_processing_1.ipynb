####1.	Create 3 sine waves with a sample rate of 8000 Hz, 2 seconds each in length with frequencies and amplitudes:
    1.110Hz at 100% amplitude
    2.220Hz at 75% amplitude

####2.	Create a white noise signal with a sample rate of 8000 Hz, 2 seconds long, with an amplitude of 500%


```python
%matplotlib inline
import numpy as np
from scipy.io import wavfile
import scipy.stats as sps
import matplotlib.pyplot as plt
import mpld3
import matplotlib.pylab as pylab
import scipy.signal as signal
plt.style.use('ggplot')
mpld3.disable_notebook()
```

```python
def create_sine(fs = 44100,dur = 3,freq = 440,A = 1.0):
    sample_len = dur * fs
    sine_samples = np.zeros(sample_len)
    T = np.linspace(0, dur, num=sample_len)
    for i in range(sample_len):
        sine_samples[i] = A * np.sin(2 * np.pi * freq * T[i])
    return sine_samples,T
```

```python
def create_whitenoise(fs = 44100,dur = 3,A = 1.0):
    sample_len = dur * fs
    sine_samples = np.zeros(sample_len)
    T = np.linspace(0, dur, num=sample_len)
    for i in range(sample_len):
        sine_samples[i] = A * np.random.uniform(-1, 1)
    return sine_samples,T
```

```python
def save_to_wav(samples,fs,path):
    samples = np.int16(samples * 32768)
    wavfile.write(path, fs, samples)
```

```python
def magnitude_spectrum_generator(signal):
    signal_dft_output = np.fft.rfft(signal)
    signal_magnitude_spectrum = [np.sqrt(i.real**2 + i.imag**2)/len(signal_dft_output) for i in signal_dft_output]
    return signal_dft_output,signal_magnitude_spectrum
```

```python
sine_110Hz_100,T_sine_110Hz_100=create_sine(fs = 8000,dur = 2,freq = 110,A = 1.0)
save_to_wav(sine_110Hz_100,8000,'110.wav')
sine_110Hz_100_dft_output,sine_110Hz_100_magnitude_spectrum = magnitude_spectrum_generator(sine_110Hz_100)

sine_220Hz_75,T_sine_220Hz_75=create_sine(fs = 8000,dur = 2,freq = 220,A = 0.75)
save_to_wav(sine_220Hz_75,8000,'220.wav')
sine_220Hz_75_dft_output,sine_220Hz_75_magnitude_spectrum = magnitude_spectrum_generator(sine_220Hz_75)

sine_440Hz_50,T_sine_440Hz_50=create_sine(fs = 8000,dur = 2,freq = 440,A = 0.50)
save_to_wav(sine_440Hz_50,8000,'440.wav')
sine_440Hz_50_dft_output,sine_440Hz_50_magnitude_spectrum = magnitude_spectrum_generator(sine_440Hz_50)

whitenoise_500,T_whitenoise_500=create_whitenoise(fs = 8000,dur = 2,A = 5.0)
save_to_wav(whitenoise_500,8000,'noise.wav')
whitenoise_500_dft_output,whitenoise_500_magnitude_spectrum = magnitude_spectrum_generator(whitenoise_500)
```

freq_detect take sample magnitude spectrum and return the freqencys that were detected.

```python
def freq_detect(comb_samples_magnitude_spectrum, threshsize=5):
    freqs = np.arange(0,len(comb_samples_magnitude_spectrum))
    thresh = np.mean(comb_samples_magnitude_spectrum) + (threshsize * np.std(comb_samples_magnitude_spectrum))
    comb_samples_magnitude_spectrum_thresh= (comb_samples_magnitude_spectrum > thresh) * comb_samples_magnitude_spectrum
    diff_spec = np.diff(comb_samples_magnitude_spectrum_thresh)
    diff_spec = np.roll(diff_spec,1)
    match = np.convolve(np.sign(diff_spec), [-1,1])
    match = np.roll(match, 1)
    idx = np.nonzero(match>0)[0]-2
    return freqs[idx]
print freq_detect(comb_samples_magnitude_spectrum, threshsize=5)
```
